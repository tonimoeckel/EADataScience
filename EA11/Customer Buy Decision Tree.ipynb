```python
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/tonimoeckel/EADataScience/master/EA11/split_buy_pc_normal.csv",delimiter=';')
df
```

```python
customer_classifier = tree.DecisionTreeClassifier()
customer_classifier.fit(df[["age", "income", "student", "credit_rating"]], df["buys"])

customer_classifier
```

```python
# data set 1st observation
test_features_0 = [[df["age"][0], df["income"][0], df["student"][0], df["credit_rating"][0]]]
predicted_features_0 = customer_classifier.predict(test_features_0)
print('Actual buy: {}; Predicted buy: {}'.format(df["buys"][0], predicted_features_0[0]))
```

```python
# data set 2nd observation
test_features_1 = [[df["age"][1], df["income"][1], df["student"][1], df["credit_rating"][1]]]
predicted_features_1 = customer_classifier.predict(test_features_1)
print('Actual buy: {}; Predicted buy: {}'.format(df["buys"][1], predicted_features_1[0]))
```

```python
with open("customer_classifier.txt", "w") as f:
    f = tree.export_graphviz(customer_classifier, out_file=f)
```
